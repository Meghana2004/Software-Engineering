```python
#hardcoded variables
def temperature_modeling(a, b, c, time):
    temperature = a * time**2 + b * time + c
    return temperature

def calculate_temperature_with_hardcoded_values():
    a_hardcoded, b_hardcoded, c_hardcoded = 1, 2, 10
    time_hardcoded = 5 
    temperature_result = temperature_modeling(a_hardcoded, b_hardcoded, c_hardcoded, time_hardcoded)
    return temperature_result, time_hardcoded

result, time_used = calculate_temperature_with_hardcoded_values()

print("Hard-coded Variables for Weather Modeling\n")
print("Temperature for hardcoded coefficients at time", time_used, "hours:", result)
print("\n")
```

```python
#keyboard input
def temperature_modeling(a, b, c, time):
    temperature = a * time**2 + b * time + c
    return temperature

def calculate_temperature_with_keyboard_input():
    print("Keyboard Input for Weather Modeling")
    a_keyboard = float(input("Enter coefficient a: "))
    b_keyboard = float(input("Enter coefficient b: "))
    c_keyboard = float(input("Enter coefficient c: "))
    time_keyboard = float(input("Enter time in hours: "))
    return temperature_modeling(a_keyboard, b_keyboard, c_keyboard, time_keyboard), time_keyboard

result, time_used = calculate_temperature_with_keyboard_input()

print("Temperature for keyboard input coefficients at time", time_used, "hours:", result)
print("\n")
```

```python
#read from a file
def temperature_modeling(a, b, c, time):
    temperature = a * time**2 + b * time + c
    return temperature

def read_coefficients_from_file(file_path):
    with open(file_path, 'r') as file:
        lines = file.readlines()
        a_file, b_file, c_file, time_file = map(float, lines[0].split())
    return a_file, b_file, c_file, time_file

file_path = r"C:\web page\weather.txt"  # Replace this with your file path
a_file, b_file, c_file, time_file = read_coefficients_from_file(file_path)

print("Read from a File for Weather Modeling\n")
print("Temperature for file input coefficients at time", time_file, "hours:", 
      temperature_modeling(a_file, b_file, c_file, time_file))
print("\n")
```

```python
#Multiple set User Input for Weather Modeling
import matplotlib.pyplot as plt
def temperature_modeling(a, b, c, time):
    temperature = a * time**2 + b * time + c
    return temperature
print("Multiple set User Input for Weather Modeling\n")
num_sets = int(input("Enter the number of sets of coefficients: "))
print("\n")
for _ in range(num_sets):
    coefficients = input("Enter coefficients (a, b, c) separated by spaces: ").split()
    a_user, b_user, c_user = map(float, coefficients)
    time_user = float(input("Enter time in hours: "))
    temperature_result = temperature_modeling(a_user, b_user, c_user, time_user)
    print("Temperature for user-input coefficients at time", time_user, "hours:", temperature_result)
    print("\n")
    time_values = range(int(time_user) + 1) 
    temperature_values = [temperature_modeling(a_user, b_user, c_user, t) for t in time_values]
    plt.plot(time_values, temperature_values, label=f'Set {_ + 1}')

plt.xlabel('Time (hours)')
plt.ylabel('Temperature')
plt.title('Temperature Modeling for Multiple Sets')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#Single set User Input
import matplotlib.pyplot as plt
def temperature_modeling(a, b, c, time):
    temperature = a * time**2 + b * time + c
    return temperature
print("User Input for Weather Modeling\n")
coefficients = input("Enter coefficients (a, b, c) separated by spaces: ").split()
a_user, b_user, c_user = map(float, coefficients)
time_user = float(input("Enter time in hours:"))
temperature_result = temperature_modeling(a_user, b_user, c_user, time_user)
print("Temperature for user-input coefficients at time", time_user, "hours:", temperature_result)
time_values = range(int(time_user) + 1) 
temperature_values = [temperature_modeling(a_user, b_user, c_user, t) for t in time_values]

plt.plot(time_values, temperature_values, label='Temperature Curve')
plt.xlabel('Time (hours)')
plt.ylabel('Temperature')
plt.title('Temperature Modeling for User Input Coefficients')
plt.legend()
plt.grid(True)
plt.show()
```
